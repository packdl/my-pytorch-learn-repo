```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('raw_data.txt')
```

```python
df = df[['epoch','model','eval_loss','eval_bleu','train_loss','train_bleu']]
```

```python
model1, model2 = df[df.model==128], df[df.model==256]
```

```python
fig, ax = plt.subplots()
ax.plot(model1.epoch, model1.eval_loss, label='val_loss')
ax.plot(model1.epoch, model1.train_loss, label='train_loss')
ax.set_xlabel('epochs')
ax.set_ylabel('loss')
ax.set_title('Model 1 (128) Loss')
ax.legend()
ax.grid(True)
```

```python
fig, ax = plt.subplots()
ax.plot(model2.epoch, model2.eval_loss, label='val_loss')
ax.plot(model2.epoch, model2.train_loss, label='train_loss')
ax.set_xlabel('epochs')
ax.set_ylabel('loss')
ax.set_title('Model 2 (256) Loss')
ax.legend()
ax.grid(True)
```

```python
fig, ax = plt.subplots()
ax.plot(model1.epoch, model1.eval_bleu, label='val_bleu')
ax.plot(model1.epoch, model1.train_bleu, label='train_bleu')
ax.set_xlabel('epochs')
ax.set_ylabel('bleu')
ax.set_title('Model 1 (128) Bleu')
ax.legend()
ax.grid(True)
```

```python
fig, ax = plt.subplots()
ax.plot(model1.epoch, model2.eval_bleu, label='val_bleu')
ax.plot(model1.epoch, model2.train_bleu, label='train_bleu')
ax.set_xlabel('epochs')
ax.set_ylabel('bleu')
ax.set_title('Model 2 (256) Bleu')
ax.legend()
ax.grid(True)
```
